```python

def get_firstname(string):
    
    first=""
    
    if len(string.split(" ")) >=2:        
        first=string.split(" ")[0]
    
    return first



######################



def get_middle(string):
    
    middle=""
    
    if len(string.split(" ")) >=3:        
        middle=string.split(" ")[1]
    
    return middle



######################


def get_lastname(string):
    
    last=string.split(" ")[-1]
    
    return last



######################


```

```python
threshold  = 0.9  # to accept a matching of two university names
   
    
# given two dataframes, i go over all names in one (linkedin)   and find  matches from the other

master_dict_linkedin_index_GS_index={}
cont=0
print "entering a loop of",len(df_1st_merge), "iters (linkedin-wos rows)......\n"


cont_ambig=0
# for row in tqdm_nofull_nametebook(df_linkedin_test.head(5000).iterrows()):       
    
lista_linkedin_idx=[]
#for row in tqdm_notebook(f_1st_merge.head(5000).iterrows()):       
for row in tqdm_notebook(df_1st_merge.iterrows()):       
    cont +=1

    linkedin_index=row[0]
       #print (cont, linkedin_index)

    lista_linkedin_idx.append(linkedin_index)

    master_dict_linkedin_index_GS_index[linkedin_index]=[]
    

    univ_linkedin=row[1].University_linkedin    
    lastname_linkedin=row[1].lastname_linkedin   
    full_name_linkedin=row[1].full_name_linkedin


    lista_posibles_GS_index=[] 


    try:   # if the whole row is not just NANs    

         
         first_name_linkedin=row[1].firstname_linkedin


         perfect_selection=df_gs[df_gs.full_name == full_name_linkedin]
         if len( perfect_selection ) ==1:
#                 print "perfect match on full name for linkedin-wos_idx!"
#                 print  perfect_selection
#                 print row[1]
#                 raw_input()
                select_df_GS_one_lastname = perfect_selection

         else:                      
              
             ### for a given last name, i only look at the GS rows with the same lastname
            pre_select_df_GS_one_lastname=df_gs[df_gs.lastname == lastname_linkedin]
            
            if len(pre_select_df_GS_one_lastname)>0:        
                select_df_GS_one_lastname = pre_select_df_GS_one_lastname[pre_select_df_GS_one_lastname.firstname ==  first_name_linkedin]

                # the selected df keeps the same indices from de original df !!!!!           
                

            else:
                select_df_GS_one_lastname = pre_select_df_GS_one_lastname


              
#                 if len(select_df_GS_one_lastname)>0:
#                     print "linkedin-wos:", linkedin_index, first_name_linkedin, lastname_linkedin, univ_linkedin
                
#                     print "size pre-selection:", len(pre_select_df_GS_one_lastname),  "   size selection:", len(select_df_GS_one_lastname)
        
#                     raw_input()


         for fila in  select_df_GS_one_lastname.iterrows():  # one row per author

            univ_GS=fila[1].University                           
            GS_index=fila[0]


            score = difflib.SequenceMatcher(None,univ_linkedin, univ_GS).ratio()  
   
  
            if score > threshold:  #if the result of the matching is worth considering               
                master_dict_linkedin_index_GS_index[linkedin_index].append(GS_index)   
               # print master_dict_linkedin_index_GS_index[linkedin_index]
                
               

                lista_posibles_GS_index.append(GS_index)   




    #      print (lista_posibles_GS_index)
         if len(lista_posibles_GS_index)>1:
                cont_ambig +=1
    except TypeError:   # for the few empty rows (all Nan)
            pass
    
    
print "done"


print "# of linkedin names with wos ambiguity:", cont_ambig
  
    
#path_merge_win='C:\\Users\\julietta\\Work\\Dropbox_studies\\Data\\Merged_LinkedIn_WoS\\'
path_merge_linux='/home/juliaponcela/at_NICO/Dropbox_collaboration_patterns/Data/Merged_LinkedIn_WoS/'


filename_pickle=path_merge_linux+"master_dict_linkedin-wos_index_GS_index.pickle"
pickle.dump(master_dict_linkedin_index_GS_index, open(filename_pickle, 'wb'))
print "written:", path+'master_dict_linkedin-wos_index_GS_index.pickle'
 
    
```
