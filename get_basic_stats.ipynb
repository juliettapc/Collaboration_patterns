```python

def get_median_num_publ(input_df):
    
    median_num_publ=np.nan            
    
    try:   
        min_num_publ=float(input_df.group_total_publ.split("-")[0])
        max_num_publ=float(input_df.group_total_publ.split("-")[1])       
    except : pass  ## if NANs

         
    try:         
        median_num_publ=int(max_num_publ-min_num_publ)/2.                   
    except: pass
    
    
    return median_num_publ
        
##################
    
    
    
def get_median_num_last_auth(input_df):
    
    median_num_last_auth=np.nan            
    
    try:   
        min_num_last_auth=float(input_df.group_num_papers_last.split("-")[0])
        max_num_last_auth=float(input_df.group_num_papers_last.split("-")[1])
    except : pass  ## if NANs

     
    
    try:                 
        median_num_last_auth=int((max_num_last_auth-min_num_last_auth)/2.)
    except : pass  ## if NANs
         
    
    
    return median_num_last_auth
            

%time df['median_num_publ'] = df.apply(lambda row: get_median_num_last_auth(row),axis=1)
%time df['median_num_last_auth'] = df.apply(lambda row: get_median_num_last_auth(row),axis=1)


print "done"
```

```python

#to get the folder entropy

def gini(list_of_values):
    sorted_list = sorted(list_of_values)
    height, area = 0, 0
    for value in sorted_list:
        height += value
        area += height - value / 2.
    fair_area = height * len(list_of_values) / 2.
    try:
        return (fair_area - area) / fair_area
    except ZeroDivisionError:
#         print "problems with:",list_of_values   # if lista=[0,0,0]  or [0]
#         raw_input()
        return np.nan




def grouping(input_df,dict_folder_id_folder_entropy):
    
   
#     folder_id=input_df.folder_id.iloc[0]
#     print folder_id
   
#     lastname=input_df.author_name.iloc[0].split(",")[0].capitalize()
    
    list_edits=list(input_df.num_edits)
    list_adds=list(input_df.num_adds)
    list_deletes=list(input_df.num_deletes)
    
    list_all_activity=replace_nans_by_zeros(list(np.array(list_edits)+np.array(list_adds)+np.array(list_deletes)))  # according to Rebecca, NANs=0 in activity columns
    
   # print list_edits, list_adds, list_deletes, "  adding arrays:", list_all_activity
    
   
    
    GINI=gini(list_all_activity)
    dict_folder_id_folder_entropy[folder_id]=GINI
    #print GINI
    #print 
    #print dict_folder_id_folder_entropy
    #raw_input()
    
 


dict_folder_id_folder_entropy={}
for folder_id in df['folder_id'].unique():
    df_selection= df[df['folder_id'] == folder_id]
    grouping(df_selection,dict_folder_id_folder_entropy)
    #print folder_id

df.head(100)
print "done"

```
